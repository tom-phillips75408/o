 IR PRACTICAL 2 :  Retrieval of documents using inverted files.

Step 1: Import required libraries

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from collections import defaultdict
```

Step 2: Download NLTK data

```python
nltk.download('punkt')
nltk.download('stopwords')
```

 Step 3: Define sample documents

```python
documents = [
    "The sky is blue and beautiful.",
    "Love this blue and beautiful sky!",
    "The quick brown fox jumps over the lazy dog.",
    "A kingâ€™s breakfast has sausages, ham, bacon, eggs, toast and beans.",
    "I love green eggs, ham, sausages and bacon!",
    "The brown fox is quick and the blue dog is lazy!",
    "The sky is very blue and the sky is very beautiful today.",
    "The dog is lazy but the brown fox is quick!"
]
```

Step 4: Define stop words

```python
stop_words = set(stopwords.words('english'))
```

Step 5: Preprocessing function

```python
def preprocess(doc):
    tokens = word_tokenize(doc.lower())               # Lowercase and tokenize
    tokens = [word for word in tokens if word.isalpha()]  # Remove punctuation
    tokens = [word for word in tokens if word not in stop_words]  # Remove stopwords
    return tokens
```

Step 6: Preprocess all documents

```python
preprocessed_docs = [preprocess(doc) for doc in documents]
```

Step 7: Build inverted index

```python
inverted_index = defaultdict(set)
for idx, tokens in enumerate(preprocessed_docs):
    for token in tokens:
        inverted_index[token].add(idx)
```

```python
inverted_index = dict(inverted_index)
```

Step 8: Search function

```python
def search(query):
    query_tokens = preprocess(query)
    if not query_tokens:
        return []

    result_sets = [inverted_index.get(token, set()) for token in query_tokens]
    matched_docs = set.intersection(*result_sets) if result_sets else set()
    return sorted(matched_docs)
```

Step 9: Interactive search loop

```python
while True:
    query = input("\nEnter your search query (or 'exit' to quit): ")
    if query.lower() == 'exit':
        break

    results = search(query)
    if results:
        print(f"\nDocuments matching '{query}':")
        for doc_id in results:
            print(f"Doc {doc_id}: {documents[doc_id]}")
    else:
        print(f"\nNo documents found for query: '{query}'")
```
